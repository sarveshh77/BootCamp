```python
import pandas as pd
d=pd.read_csv('Students.csv')
```

```python
print(d.head())
```

```python
print(d.isnull().sum())
```

```python
d['MathScore'].fillna(d['MathScore'].mean(),inplace=True)
```

```python
d
```

```python
d['EnglishScore'].fillna(d['EnglishScore'].mean(),inplace=True)
```

```python
d
```

```python
from sklearn.preprocessing import LabelEncoder

l=LabelEncoder()

l.fit(d['Gender'])
```

```python
d["Gender"]=l.transform(d[["Gender"]])
```

```python
d
```

```python
d['TotalScore']=d['MathScore']+d['EnglishScore']
```

```python
d
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
d[['MathScore','EnglishScore','TotalScore']]=ss.fit_transform(
d[['MathScore','EnglishScore','TotalScore']])
```
